PRECISION
Determina de todos los que el modelo predijo como positivos (ej. malignos) ¿Cuántos eran realmnete positivos?

RECALL
Se considera mejor para datos médicos, en especial para cancer 
De todos los positivos reales, ¿cuántos detecto el modelo?

```python
# Accuracy = (TP + TN) / (TP + TN + FP + FN)
# Precision = TP / (TP + FP)
# Recall = TP / (TP + FN)
# F1 = 2 · (Precision · Recall) / (Precision + Recall)
```

F1 - Score
Balance entre Precisión y Recall

# MODELO DE REGRESIÓN LOGÍSTICA 

Se basa en la función sigmoide, que transforma cualquier valor real en una probabilidad entre 0 y 1

```python
#Graficando curva sigmoide

import numpy as np
import matplotlib.pyplot as plt

# Definir la función sigmoide
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Crear rango de valores
x = np.linspace(-10, 10, 100)
y = sigmoid(x)

# Graficar
plt.figure()
plt.plot(x, y)
plt.axhline(0.5)   # línea del threshold
plt.axvline(0)     # punto central
plt.title("Curva Sigmoide")
plt.xlabel("z")
plt.ylabel("Probabilidad")
plt.show()

```

# METRICAS DE ERROR
MAE ---> Nos dará el promedio del valor absoluto entre el valor real y la predicción
RMSE ---> Nos dará la raíz del promedio del error cuadrático

COMPARACION DE RESULTADOS ANTES Y DESPUES DEL ESCALADO

```python
#Entrenar ambos modelos 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# -------- KNN SIN ESCALADO --------
knn_raw = KNeighborsClassifier(n_neighbors=5)
knn_raw.fit(X_train, y_train)
y_pred_raw = knn_raw.predict(X_test)

# Métricas sin escalado
acc_raw = accuracy_score(y_test, y_pred_raw)
prec_raw = precision_score(y_test, y_pred_raw)
rec_raw = recall_score(y_test, y_pred_raw)
f1_raw = f1_score(y_test, y_pred_raw)


# -------- KNN CON ESCALADO --------
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsClassifier(n_neighbors=5))
])

pipeline.fit(X_train, y_train)
y_pred_scaled = pipeline.predict(X_test)

# Métricas con escalado
acc_scaled = accuracy_score(y_test, y_pred_scaled)
prec_scaled = precision_score(y_test, y_pred_scaled)
rec_scaled = recall_score(y_test, y_pred_scaled)
f1_scaled = f1_score(y_test, y_pred_scaled)
```

# ENSEMBLE METHODS

Con la idea de entender el funcionamiento del ensamble, voy a desmenuzar un poco más su proceso, para saber: qeu hace cada árbol, como votan y como mejoran el rendimiento 


```python
#Modelo base con sólo un árbol

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

tree = DecisionTreeClassifier(random_state=42)
tree.fit(X_train, y_train)

y_pred_tree = tree.predict(X_test)

print("Árbol individual Accuracy:", accuracy_score(y_test, y_pred_tree))
```

```python
#Entrenamiento de varios árbooles (idea del ensamble)

import numpy as np

n_models = 10
predictions = []

for i in range(n_models):
    X_sample = X_train.sample(frac=1, replace=True, random_state=i)
    y_sample = y_train.loc[X_sample.index]
    
    model = DecisionTreeClassifier(random_state=i)
    model.fit(X_sample, y_sample)
    
    preds = model.predict(X_test)
    predictions.append(preds)

predictions = np.array(predictions)
```

Con esto tomamos en cuenta 10 modelos x predicciones sobre el test

```python
#Voto mayoritario, en clasificación es la que más se repite entre los modelos 

import numpy as np

# Voto mayoritario manual
y_pred_ensemble = []

for i in range(predictions.shape[1]):  # recorrer cada muestra del test
    votos = predictions[:, i]
    valores, conteos = np.unique(votos, return_counts=True)
    y_pred_ensemble.append(valores[np.argmax(conteos)])

y_pred_ensemble = np.array(y_pred_ensemble)

print("Ensemble manual Accuracy:", accuracy_score(y_test, y_pred_ensemble))
```

Se miran los votos, ( en este caso 10) , cuenta cuantas veces aparece cada clase y se queda con el que más se repite.

# RANDOM SEARCH

Prueba combinaciones aleatorias, es mucho más eficiente y encuentra casi el mismo resultado en menos tiempo (en comparación con Grid Search)

```python
#Implementación con Gradient Boosting
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import StratifiedKFold
import numpy as np

gb = GradientBoostingClassifier(random_state=42)

param_dist = {
    'n_estimators': np.arange(100, 500, 50),
    'learning_rate': np.linspace(0.01, 0.2, 20),
    'max_depth': np.arange(2, 6),
    'subsample': np.linspace(0.6, 1.0, 5),
    'min_samples_split': np.arange(2, 10),
    'min_samples_leaf': np.arange(1, 5)
}

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

random_search = RandomizedSearchCV(
    estimator=gb,
    param_distributions=param_dist,
    n_iter=30,              # número de combinaciones aleatorias
    scoring='f1',
    cv=skf,
    verbose=1,
    random_state=42,
    n_jobs=-1
)

random_search.fit(X_train, y_train)
```

```python
#Obteniendo el mejor modelo 
best_gb_random = random_search.best_estimator_

print("Mejores parámetros:")
print(random_search.best_params_)

print("Mejor F1 CV:")
print(random_search.best_score_)
```
